```python
import os
import cv2
import matplotlib.pyplot as plt
import  face_recognition
import PIL.Image
import PIL.ImageDraw
import pandas as pd
from openpyxl.workbook import Workbook
```

```python
path="Faces"
images=os.listdir(path)
j=-1
print(images)
for img in images:
    img="Faces/"+img
    j=j+1
    images[j]=img

cols= ["left_A","left_B","left_C","left_ear","right_A","right_B","right_C","right_ear","AVG_EAR","Class"]
df = pd.DataFrame([], columns = cols)
```

```python
from scipy.spatial import distance as dist
i=0 
cls=0
for image in images:
    input_image = face_recognition.load_image_file(image)
    face_landmarks = face_recognition.face_landmarks(input_image)
    landmark1 = face_landmarks[0].get('left_eye')
    landmark2 = face_landmarks[0].get('right_eye')
    left_A = dist.euclidean(landmark1[1], landmark1[5])
    left_B = dist.euclidean(landmark1[2], landmark1[4])
    left_C = dist.euclidean(landmark1[0], landmark1[3])
    left_ear = (left_A + left_B) / (2.0 * left_C)
    right_A = dist.euclidean(landmark2[1], landmark2[5])
    right_B = dist.euclidean(landmark2[2], landmark2[4])
    right_C = dist.euclidean(landmark2[0], landmark2[3])
    right_ear = (right_A + right_B) / (2.0 * right_C)
    AVG_EAR= ( right_ear + left_ear )/2
    if(AVG_EAR<0.25):
        cls=0
    else: cls=1
    out=[left_A,left_B,left_C,left_ear,right_A,right_B,right_C,right_ear,AVG_EAR,cls]
    i=i+1
    df.loc[i]=out
```
